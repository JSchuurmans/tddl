```python
import pandas as pd
```

```python
paths = {
    'C10': {
        'RN18': "./tables/kt_rn18_c10.zip",
        'Gar': "./tables/kt_gar_c10.zip",
    },
    'FM': {
        'RN18': "./tables/kt_rn18_fm.zip",
        'Gar': "./tables/kt_gar_fm.zip",
    }
}

dicts = {
    k: {k2:pd.read_pickle(v2) for k2,v2 in v.items()} 
    for k,v in paths.items()
}

kt = pd.concat({k:pd.concat(v) for k,v in dicts.items()}, axis=0, names=['data', 'model','acc.'])
kt
```

```python
paths = {
    'C10': {
        'RN18': "./tables/kt_rn18_c10_classification_error.zip",
        'Gar': "./tables/kt_gar_c10_classification_error.zip",
    },
    'FM': {
        # 'RN18': "./tables/kt_rn18_fm.zip",
        'Gar': "./tables/kt_gar_fm_classification_error.zip",
    }
}

dicts = {
    k: {k2:pd.read_pickle(v2) for k2,v2 in v.items()} 
    for k,v in paths.items()
}

kt = pd.concat({k:pd.concat(v) for k,v in dicts.items()}, axis=0, names=['data', 'model','acc.'])
kt
```

```python
neat_table_indices = {
    'log_valid_error_before_ft': "V B",
    'log_valid_error': "V A",
    'log_test_error': "T A",
}
neat_table_columns = {				
    "relative_norm_weight": "RW",
    "scaled_norm_weight": "SW",
    "diff_norm_weight": "DW",
    "relative_norm": "RF",
    "scaled_norm": "SF",
    "norm_diff": "DF",
    "layers": "L",
}
kt.rename(index=neat_table_indices ,columns=neat_table_columns, inplace=True)
kt
```

```python
d = {
    'W': kt[['RW','SW','DW']],
    'F': kt[['RF','SF','DF']],
    'L': kt[['L']],
}
df = pd.concat(d,axis=1,
    # names=['calculated on','metric']
)
df
```

```python
# bar_paths = [
#     "tables/kt_rn18_c10_bar_log_error.zip",
#     # "tables/kt_rn18_fm_bar.zip",
#     "tables/kt_gar_c10_bar_log_error.zip",
#     "tables/kt_gar_fm_bar_log_error.zip",
# ]

bar_paths = [
    "tables/kta_rn18_c10_bar_log_error.zip",
    "tables/kta_gar_c10_bar_log_error.zip",
    "tables/kta_gar_fm_bar_log_error.zip",
]
```

```python
bars = pd.DataFrame()
for path in bar_paths:
    bars = bars.append(pd.read_pickle(path))


bars.drop(bars.index[bars["approx_type"] == "layers"], inplace=True)
bars
```

```python
bars['model & dataset'] = bars['model'] +" & "+ bars['dataset']
```

```python
replacements = {
    'valid_acc_before_ft': "Validation accuracy before fine-tuning", 
    'valid_acc': "Validation accuracy after fine-tuning",
    'test_acc': "Test accuracy after fine-tuning", 
    'relative_norm_weight': 'Relative Weights',
    'scaled_norm_weight': 'Scaled Weights',
    'diff_norm_weight': 'Absolute Weights',
    'relative_norm': 'Relative Features',
    'scaled_norm': 'Scaled Features',
    'norm_diff': 'Absolute Features',
    'log_valid_error_before_ft': 'Validation error before fine-tuning',
    'log_valid_error': 'Validation error after fine-tuning',
    'log_test_error': 'Test error after fine-tuning'
    # 'log_norm_diff': 'Log norm of the Absolute between training features',
}
bars_neat = bars.replace(replacements)
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

g = sns.catplot(
    data=bars_neat, kind="bar",
    x='approx_type',
    y='kt',
    hue='acc_type',
    ci="sd", palette="dark", alpha=.6, height=6,
    col="model & dataset",
    # row="model", 
)
g.despine(left=True)
g.set_axis_labels("", "Kendall's tau")
# g.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
#           fancybox=True, shadow=True, ncol=5)
g.legend.set_title("")
for i in range(1):
    for j in range(3):
        gax = g.axes[i][j]
        # xlabel = gax.get_xlabel()
        # ylabel = gax.get_ylabel()
        # print(xlabel)
        # if xlabel in replacements.keys():
        #     print(xlabel)
        #     g.axes[i][j].set_xlabel(replacements[xlabel])
        gax.set_xticklabels(gax.get_xticklabels(),rotation = 45, fontdict={'horizontalalignment':'right'})# 'center', 'left', 'right')

sns.move_legend(g, "lower center", bbox_to_anchor=(.4, 1), ncol=3, title=None, frameon=False)

# g.savefig("../../plots/kt_bar_errors_performances")
```
