```python
import pandas as pd
```

```python
paths = {
    'ResNet-18 - CIFAR-10': "./tables/kt_rn18_c10.zip",
    'ResNet-18 - F-MNIST': "./tables/kt_rn18_fm.zip",
    'GaripovNet - CIFAR-10': "./tables/kt_gar_c10.zip",
    'GaripovNet - F-MNIST': "./tables/kt_gar_fm.zip",
}

d = {k:pd.read_pickle(v) for k,v in paths.items()}
pd.concat(d, axis=0,names=['model - data combination','accuracy'])
```

```python
paths = {
    'C10': {
        'RN18': "./tables/kt_rn18_c10.zip",
        'Gar': "./tables/kt_gar_c10.zip",
    },
    'FM': {
        'RN18': "./tables/kt_rn18_fm.zip",
        'Gar': "./tables/kt_gar_fm.zip",
    }
}

dicts = {
    k: {k2:pd.read_pickle(v2) for k2,v2 in v.items()} 
    for k,v in paths.items()
}

kt = pd.concat({k:pd.concat(v) for k,v in dicts.items()}, axis=0, names=['data', 'model','acc.'])
kt
```

```python
neat_table_indices = {
    'valid_acc_before_ft': "V B",
    'valid_acc': "V A",
    'test_acc': "T A",
}
neat_table_columns = {				
    "relative_norm_weight": "RW",
    "scaled_norm_weight": "SW",
    "diff_norm_weight": "DW",
    "relative_norm": "RF",
    "scaled_norm": "SF",
    "norm_diff": "DF",
    "layers": "L"
}
kt.rename(index=neat_table_indices ,columns=neat_table_columns, inplace=True)
kt
```

```python
d = {
    'W': kt[['RW','SW','DW']],
    'F': kt[['RF','SF','DF']],
    'L': kt[['L']],
}
df = pd.concat(d,axis=1,
    # names=['calculated on','metric']
)
df
```
